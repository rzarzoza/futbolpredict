Import all the libraries to analyze the football data

```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv = "../data/all_matches.csv"
df = pd.read_csv(csv, low_memory=False)
```

```python
counts = df["FTR"].value_counts()
```

```python
plt.bar(counts.index, counts.values, color=["green", "gray", "red"])
plt.title("Results Distribution")
plt.xlabel("Result")
plt.ylabel("Number of matches")
plt.show()
```

```python
with open("columns.txt", "w") as file:
    for col in df.columns:
        file.write(col)
        file.write(" \n")
```
